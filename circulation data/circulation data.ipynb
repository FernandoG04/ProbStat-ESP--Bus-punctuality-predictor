```python
import pandas as pd
from geopy.geocoders import Nominatim
import folium

# Load the data
df = pd.read_csv('comptages_vehicules_cyclistes_pietons.csv')

# Drop all columns except 'Nom_intersection', 'Latitude', 'Longitude'
df = df[['Nom_Intersection', 'Latitude', 'Longitude', 'Id_Intersection']]

df['Id_Intersection'] = df['Id_Intersection'].astype(str)

# Drop duplicates and get the total number of rows
df = df.drop_duplicates()

#Create a dictionary with dataframes that have a Nom_Intersection that contains a certain string (street name) each
streets_dict = {}
streets = ['Sauvé', 'Pie-IX', 'Crémazie', 'Saint-Michel', 'René-Lévesque', 'Parc', 'Champagneur']
for street in streets:
    streets_dict[street] = df[df['Nom_Intersection'].str.contains(street)]

print("Total number of rows: ", len(df))
df.head()
```

```python
# Create a list of colors
colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred',
          'lightred', 'beige', 'darkblue', 'darkgreen', 'cadetblue',
          'darkpurple', 'white', 'pink', 'lightblue', 'lightgreen',
          'gray', 'black', 'lightgray']

# Create a dictionary mapping each street to a color
color_dict = {street: colors[i % len(colors)] for i, street in enumerate(streets_dict)}

# Create the map
m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=13)

# Add the pins
for street in streets_dict:
    for _, row in streets_dict[street].iterrows():
        folium.Marker([row['Latitude'], row['Longitude']], 
                      popup=row['Nom_Intersection'] + ', ' + row['Id_Intersection'], 
                      icon=folium.Icon(color=color_dict[street])).add_to(m)

# Display the map
m.save('map2.html')
m
```
